## Checking CDF

```python
from scipy.stats import rv_histogram

def cdf(T):
    return (2 / np.pi) * np.arcsin(np.sqrt(T))

def compute_cdf_scipy(data, num_bins=10):
    """
    Compute the empirical cumulative distribution function (CDF).
    
    Parameters:
    - data: NumPy array of observations.
    - num_bins: Number of bins to use for the histogram (default is 10).
    
    Returns:
    - x_vals: The sorted array of observations (bin centers).
    - cdf_vals: The corresponding CDF values.
    """
    # Create a histogram and normalize it to get a probability distribution
    hist, bin_edges = np.histogram(data, bins=num_bins, density=True)
    
    # Create a scipy rv_histogram object
    hist_dist = rv_histogram((hist, bin_edges))
    
    # Compute the CDF using the fitted histogram distribution
    x_vals = np.linspace(np.min(data), np.max(data), 1000)
    cdf_vals = hist_dist.cdf(x_vals)
    
    return x_vals, cdf_vals

tv100, cdf100   = compute_cdf_scipy(t1_colloidal_post_tf100)  # change these all at once for various plots
tv250, cdf250   = compute_cdf_scipy(t1_colloidal_post_tf250)  #
tv500, cdf500   = compute_cdf_scipy(t1_colloidal_post_tf500)  #
tv750, cdf750   = compute_cdf_scipy(t1_colloidal_post_tf750)  #
tv1000, cdf1000 = compute_cdf_scipy(t1_colloidal_post_tf1000) # 

# Optionally, visualize the CDF
import matplotlib.pyplot as plt
plt.plot(np.linspace(0,1,100),cdf(np.linspace(0,1,100)),color='orange',linewidth=2.5,label='analytical') #,label='$2/\pi$ $\\arcsin(\sqrt{x})$'
plt.plot(tv100, cdf100,label='$t_f$ = 100')
plt.plot(tv250, cdf250,label='$t_f$ = 250')
plt.plot(tv500, cdf500,label='$t_f$ = 500')
plt.plot(tv750, cdf750,label='$t_f$ = 750')
plt.plot(tv1000, cdf1000,label='$t_f$ = 1000')
plt.legend(loc=(1.04, 0.25))
plt.xlabel('$\mathcal{T}_1$',fontsize=16)
#plt.xlabel('$\mathcal{T}_2$',fontsize=16)
plt.ylabel('Cumulative Distribution')
```
